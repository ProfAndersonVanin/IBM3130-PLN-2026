# **Exemplo de PLN utilizando um Corpus Real**
**Livro**: Dom Casmurro (Machado de Assis)

# **1. Baixando o texto**

```python
import requests
url = "https://www.gutenberg.org/cache/epub/55752/pg55752.txt"
texto = requests.get(url).text
print(texto[:1500])
```

# **2. Selecionando um trecho do livro**

Para a aula não é necessário utilizar o livro inteiro.

```python
inicio = 15000
fim = 21000

capitulo = texto[inicio:fim]
print(capitulo)
```

# **3. Tokenização**

```python
import nltk
nltk.download("punkt_tab")
from nltk.tokenize import word_tokenize
tokens = word_tokenize(capitulo, language="portuguese")

print("Quantidade de tokens:", len(tokens))
print(tokens[:50])
```

# **4. Normalização**

```python
import string
tokens_normalizados = []
for palavra in tokens:
    palavra = palavra.lower()
    if palavra not in string.punctuation:
        tokens_normalizados.append(palavra)
print(tokens_normalizados[:50])
```

# **5. Stopwords**

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
stop = stopwords.words("portuguese")
tokens_sem_stop = []
for palavra in tokens_normalizados:
    if palavra not in stop:
        tokens_sem_stop.append(palavra)
print(tokens_sem_stop[:50])
```

# **6. Stopwords customizadas**

Suponha que queremos remover nomes próprios do livro.

```python
stop_customizada = [
    "bentinho",
    "capitu",
    "escobar",
    "dona",
    "josé"
]

resultado = []
for palavra in tokens_sem_stop:
    if palavra not in stop_customizada:
        resultado.append(palavra)
print(resultado[:50])
```

# **7. Stemming**

```python
nltk.download("rslp")
from nltk.stem import RSLPStemmer
stemmer = RSLPStemmer()
radicais = []
for palavra in resultado:
    radicais.append(
        stemmer.stem(palavra)
    )
print(radicais[:50])
```

# **9. Pipeline reutilizável com NLTK**

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer

import nltk
import string

nltk.download("punkt_tab")
nltk.download("stopwords")
nltk.download("rslp")
```

```python
def preprocessar_nltk(texto):
    tokens = word_tokenize(texto, language="portuguese")
    resultado = []
    for palavra in tokens:
        palavra = palavra.lower()
        if palavra not in string.punctuation:
            if palavra not in stop:
                radical = stemmer.stem(palavra)
                resultado.append(radical)
    return resultado
```

Aplicação

```python
saida = preprocessar_nltk(capitulo)
print(saida[:100])
```

# **10. Pipeline reutilizável com spaCy**

```python
def preprocessar_spacy(texto):
    doc = nlp(texto)
    resultado = []
    for token in doc:
        if token.is_punct == False:
            if token.is_stop == False:
                resultado.append(token.lemma_.lower())
    return resultado
```

Aplicação

```python
saida = preprocessar_spacy(capitulo)
print(saida[:100])
```

```python
print("Texto original")
print(capitulo[:400])

print()

print("Após NLTK")
print(preprocessar_nltk(capitulo)[:50])

print()

print("Após spaCy")
print(preprocessar_spacy(capitulo)[:50])
```
