## 2. Importação das bibliotecas


```python
import nltk
import spacy
import string

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import RSLPStemmer
```

## 3. Download dos recursos do NLTK

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('rslp')
```

## 4. Corpus de exemplo

Utilizaremos um pequeno corpus com frases reais relacionadas à Inteligência Artificial.

```python
corpus = [
    "A inteligência artificial está transformando a educação brasileira.",
    "Pesquisadores desenvolveram um novo algoritmo para análise de sentimentos.",
    "O processamento de linguagem natural permite interpretar textos automaticamente.",
    "Universidades utilizam aprendizado de máquina em diversos projetos."
]

for frase in corpus:
    print(frase)
```

# **Etapa 1. Tokenização**

Separar o texto em palavras.

```python
texto = corpus[0] # Aqui estamos utilizando só a primeira frase!
tokens = word_tokenize(texto, language="portuguese")
print(tokens)
```

# **Etapa 2. Normalização**

Converter todas as palavras para letras minúsculas e remover pontuações.

```python
tokens_normalizados = []
for palavra in tokens:
    palavra = palavra.lower()
    if palavra not in string.punctuation:
        tokens_normalizados.append(palavra)
print(tokens_normalizados)
```

# **Etapa 3. Stopwords**
Lista padrão do NLTK

```python
stop_padrao = stopwords.words("portuguese")
resultado = []
for palavra in tokens_normalizados:
    if palavra not in stop_padrao:
        resultado.append(palavra)
print(resultado)
```

## **Lista customizada**

Além da lista padrão, podemos criar uma lista específica para o nosso problema.

```python
stop_customizada = [
    "inteligência",
    "artificial"
]

resultado = []
for palavra in tokens_normalizados:
    if palavra not in stop_padrao:
        if palavra not in stop_customizada:
            resultado.append(palavra)
print(resultado)
```

# **Etapa 4. Stemming**

Redução das palavras ao seu radical.

```python
stemmer = RSLPStemmer()
radicais = []
for palavra in resultado:
    radical = stemmer.stem(palavra)
    radicais.append(radical)
print(radicais)
```

> **Observe que o radical nem sempre corresponde a uma palavra existente no idioma.**

# **Etapa 5. Lematização**

O spaCy retorna a forma correta da palavra existente no dicionário.

```python
nlp = spacy.load("pt_core_news_sm")
texto = corpus[0]
doc = nlp(texto)
for token in doc:
    print(token.text, "->", token.lemma_)
```

# **Etapa 6. Pipeline reutilizável com NLTK**

Agora criaremos uma função para realizar todo o pré-processamento automaticamente.

```python
stop_padrao = stopwords.words("portuguese")
stemmer = RSLPStemmer()
def preprocessar_nltk(texto):
    tokens = word_tokenize(texto, language="portuguese")
    resultado = []
    for palavra in tokens:
        palavra = palavra.lower()
        if palavra not in string.punctuation:
            if palavra not in stop_padrao:
                radical = stemmer.stem(palavra)
                resultado.append(radical)
    return resultado
```

Aplicando ao corpus:

```python
for frase in corpus:
    print(frase)
    print(preprocessar_nltk(frase))
    print()
```

# **Etapa 7. Pipeline reutilizável com spaCy**

Agora criaremos uma função equivalente utilizando o spaCy.

```python
nlp = spacy.load("pt_core_news_sm")
def preprocessar_spacy(texto):
    doc = nlp(texto)
    resultado = []
    for token in doc:
        if token.is_punct == False:
            if token.is_stop == False:
                resultado.append(token.lemma_.lower())
    return resultado
```

Aplicando ao corpus:

```python
for frase in corpus:
    print(frase)
    print(preprocessar_spacy(frase))
    print()
```

## **Comparando Stemming e Lematização**

```python
palavra = "transformando"
print("Palavra Original:", palavra)
stemmer = RSLPStemmer()
print("Stemming :", stemmer.stem(palavra))

doc = nlp(palavra)
for token in doc:
    print("Lematização:", token.lemma_)
```

- **Tokenização** ==> Divide o texto em palavras (tokens).
- **Normalização** ==> Padroniza o texto (minúsculas e remoção de pontuação).
- **Stopwords** ==> Remove palavras pouco informativas.
- **Stemming** ==> Reduz a palavra ao seu radical.
- **Lematização** ==> Obtém a forma canônica da palavra.
- **Pipeline NLTK** ==> Automatiza o pré-processamento utilizando tokenização, stopwords e stemming.
- **Pipeline spaCy**	==> Automatiza o pré-processamento utilizando tokenização, stopwords e lematização.

> **Em aplicações reais de PLN, a lematização tende a produzir resultados linguisticamente mais precisos do que o stemming, porém demanda mais recursos computacionais. Já o stemming é mais simples e rápido, sendo adequado para tarefas em que a velocidade é mais importante do que a precisão linguística.**
